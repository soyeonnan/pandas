```python
df = pd.read_csv('data/credit.csv')
df
```

```python
df.columns
```

```python
# 기초 통계
print(df['나이'].mean()) # 평균
print(df['학력'].mode()) # 최빈값
print(df['사용금액'].min()) # 최소
print(df['사용금액'].max()) # 최대
```

```python
df.describe() # 전체 통계
```

```python
# 성별 별 사용 금액과 나이, 사용 횟수 평균
df.groupby('성별').mean(numeric_only=True)
```

```python
df.groupby('성별')['나이'].mean()
```

```python
# 성별 별 아이와 사용 금액 평균
# df.groupby('성별')['나이','사용금액'].mean() ->이렇게 하면 안 나옴
df.groupby('성별')[['나이','사용금액']].mean()
```

```python
# 사용 금액이 많으면 사용 횟수도 많을까?
df[['사용금액','사용횟수']].corr()
```

```python
df[df['나이'] >= 40]
```

```python
df[ df['성별']=='M']
```

```python
df.query("성별 == 'M' and 기혼 == 'Married'").describe()
```

```python
# 결혼 한 남자들의 사용 금액 평균
a = df.query("성별 == 'M' and 기혼 == 'Married'")['사용금액'].mean()
# 결혼 안 한 남자들의 사용 금액 평균
b = df.query("성별 == 'M' and 기혼 == 'Single'")['사용금액'].mean()
print(f"결혼 한 남자들 사용 금액 평균 : {a}")
print(f"결혼 안 한 남자들 사용 금액 평균 : {b}")
a-b
```

```python
(a-b)/a*100 # 결혼 한 사람이 약 6% 더 높음
```
