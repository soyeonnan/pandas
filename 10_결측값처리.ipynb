```python
"""
결측값 : 데이터셋에 값이 없거나 기록되지 않은 것
        NaN,NULL, None, ''
결측값이 있는 상태로 분석을 하게되면.. 
- 결과가 왜곡 될 수 있음(평균 계산 -> 결측값이 개수에 포함)
- 신뢰도 하락
"""
```

```python
import pandas as pd
df = pd.read_csv('data/country_timeseries.csv')
df.head()
```

```python
df.isnull()
df.isna() # 둘 다 똑같음
```

```python
df.isnull().sum()
# 컬럼 별 결측치 갯수를 볼 수 있음
```

```python
df.isnull().sum().sum() # 전체 결측값 개수
```

```python
df.isnull().sum(axis=1) # 행 별 결측값 개수
```

```python
# 결측값을 처리
df.fillna(0) # 결측값을 전부 0으로 변경 ( 저장은 안 되는 듯 )
```

```python
df.ffill() # 앞의 값으로 채움
```

```python
df.bfill() # 뒤에 값으로 채움
```

```python
df.interpolate() # 중앙 값으로 채워짐
```

```python
df = pd.DataFrame({
  'A':[1,2,None,4,5],
  'B':[None,2,3,None,5],
  'C':[1,None,None,4,5]
})
df
```

```python
df['A'] = df['A'].fillna(df['A'].mean()) # 평균값으로 채움
df
```

```python
df['B'] = df['B'].fillna(df['B'].median()) # 중앙?값으로 채움
df
```

```python
"""
결측값을 어떻게 처리 할 것인가
데이터 수, 결측값 발생 패턴, 데이터 특성, 어떤 걸 분석하려고 하는지
ffill, bfill : 시계열데이터 (시간 순서에 따라 발생되는 데이터)
                연속적인 관측값 
                주식가격, 날씨
평균 : 데이터 분포가 균일하고, 결측값 발생이 랜덤
      (귀찮을 때)
      성적 데이터 셋에 누락 된 점수를 채울 경우
      (극단 값이 있을 경우 : 중앙값)

"""
```

```python
df = pd.DataFrame({
  'name' : ['홍길동','고길동',None,'유재석'],
  'age' : [20,None,30,25],
  'city' : ['서울','경기','부산',None]
})
df
```

```python
#  결측값이 있는 행 다 삭제
df.dropna()
```

```python
# 행에 모든 데이터가 결측값일 경우만 삭제
df.dropna(how='all')
```

```python
# 열 기준으로 결측치가 있는 열 삭제
df.dropna(axis=1)
```

```python
# age에 있는 결측치 행
df.dropna(subset=['age'])
```

```python
# 2개 이상의 결측치가 아닌 값이 있으면 삭제 안 함
df.dropna(thresh=2) # 숫자를 봐꾸면 숫자만큼 (결측치가 아닌거의 갯수)
```
