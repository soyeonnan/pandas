```python
import folium
import requests
import branca.colormap as cm
import matplotlib.pyplot as plt

# 인구수 데이터 ( 구 이름과 인구수 )
population_data = {
    '강남구': 543098, '강동구': 428549, '강북구': 311244, '강서구': 607747,
    '관악구': 497119, '광진구': 351413, '구로구': 420325, '금천구': 244654,
    '노원구': 518366, '도봉구': 328488, '동대문구': 347808, '동작구': 384215,
    '마포구': 376506, '서대문구': 316426, '서초구': 440123, '성동구': 317917,
    '성북구': 423789, '송파구': 646903, '양천구': 467117, '영등포구': 368402,
    '용산구': 243842, '은평구': 445714, '종로구': 144340, '중구': 126152, '중랑구': 401361
}

# GeoJSON 파일 로드
geojson_url = 'https://raw.githubusercontent.com/southkorea/southkorea-maps/master/kostat/2013/json/skorea_municipalities_geo_simple.json'
geojson_data = requests.get(geojson_url).json()

print( type(geojson_data) ) # 딕셔너리
```

```python
m = folium.Map( location=[37.5665, 126.09780], zoom_start=11 )
```

```python
colormap = cm.LinearColormap(
  plt.get_cmap('Pastel1').colors,
  vmin=min( population_data.values() ),
  vmax=max( population_data.values() ),
  caption="인구수"
)
```

```python
def style_function(feature):
  # 구
  gu = feature['properties']['name']

  # 인구수
  pop = population_data.get(gu, 0) # population_data[gu] -> 이거는 결측치가 None으로 나오고 .get은 0으로 나옴
  return {
    'fillColor' : colormap(pop),
    'color' : 'black',
    'weight' : 2,
    'fillOpacity' : 0.6
  }
```

```python
folium.GeoJson(
  geojson_data,
  name="서울 인구수",
  style_function=style_function
).add_to(m)
```

```python
colormap.add_to(m)
```

```python
m.save('서울인구.html')
```
