```python
df = pd.DataFrame({
  'A' : [1, 2, None, 4, 5],
  'B' : [None, 2, 3, None, 5],
  'C' : [1, None, None, 4, 5]
})
df
```

```python
df['A'] = df['A'].fillna( df['A'].mean() )
df
```

```python
df['B'] = df['B'].fillna( df['B'].median() )
df
```

```python
"""
결측값을 어떻게 처리할것인가
데이터 수, 결측값 발생 패턴, 데이터 특성, 어떤걸 분석하려고 하는지
ffill, bfill : 시계열데이터( 시간 순서에 따라 발생되는 데이터 )
               연속적인 관측값
               주식가격, 날씨
평균 : 데이터 분포가 균일하고, 결측값 발생이 랜덤
      성적 데이터셋에 누락된 점수를 채울 경우
      ( 극단 값이 있을 경우 : 중앙값 )
"""
```

```python
df = pd.DataFrame({
  'name' : ['홍길동', '고길동', None, '유재석', None],
  'age' : [20, None, 30, 25, None],
  'city' : ['서울', '경기', '부산', None, None]
})
df
```

```python
# 결측값이 있는 행 다 삭제
df.dropna()
```

```python
# 행에 모든 데이터가 결측값일 경우만 삭제
df.dropna(how='all')
```

```python
# 열 기준으로 결측치가 있는 열 삭제
df.dropna(axis=1)
```

```python
# age가 결측치인 행만 삭제
df.dropna(subset=['age'])
```

```python
# 2개 이상의 결측치가 아닌 값이 있으면 삭제안함
df.dropna(thresh=2)
```
