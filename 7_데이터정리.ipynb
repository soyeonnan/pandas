```python
ebola = pd.read_csv('data/country_timeseries.csv')
ebola.head(3)
```

```python
ebola_melt = ebola.melt(id_vars=['Date','Day'])
ebola_melt.head(3)
```

```python
var_split = ebola_melt.variable.str.split('_')
var_split
```

```python
status = var_split.str.get(0)
status
```

```python
country = var_split.str.get(1)
country
```

```python
ebola_melt['status']=status
ebola_melt['country']=country
ebola_melt.head(5)
```

```python
ebola_melt.drop(columns='variable', inplace=True)
ebola_melt.head(3)
```

```python
ebola_melt = ebola.melt(id_vars=['Date','Day'])

var_split = ebola_melt.variable.str.split('_',expand=True)
var_split
```

```python
ebola_melt[ ['status','country'] ] = var_split
ebola_melt.head(3)
```
