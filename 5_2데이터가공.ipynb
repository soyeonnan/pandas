```python
# mpg.csv을 데이터 프레임으로 가져와서 category와 cty만 추출해서 새 데이터 프레임을 생성
# 새로 생성 된 데이터 프레임을 이용해서 cotegory가 suv인 cty의 평균과 compact인 cty의 평균 중
# 어떤 게 더 높은지 구해보십쇼
df = pd.read_csv('data/mpg.csv')
df
```

```python
_df = df.loc[:,['category','cty']]
_df
```

```python
_df.info() # 타입 확인
```

```python
suv_mean = _df.query("category=='suv'")['cty'].mean()
compact_mean = _df.query("category=='compact'")['cty'].mean()

suv_mean
compact_mean

if suv_mean>compact_mean:
  print("suv가 도시 평균 연비가 더 좋습니다.")
else:
  print("compact가 도시 평균 연비가 더 좋습니다.")

print(f"suv평균연비 : {suv_mean}")
print(f"compact평균연비 : {compact_mean}")
```

```python
# 파생 변수 만들기
# df['추가 할 파생 변수 이름'] = 계산식
exam.head(2)
```

```python
# 파생변수 추가 해주는 함수
# .assign( 추가 할 변수명 = 계산식 )
exam.assign(total = exam['math'] + exam['english'] + exam['science'])
```

```python
exam.assign(
  total = exam['math'] + exam['english'] + exam['science'],
  mean = (exam['math'] + exam['english'] + exam['science']) / 3 
)
```

```python
# 함수/메서드 체이닝 (여러 가지 함수를 체인마냥 이어서 작성 가능)
# test를 추가하고 math 기준으로 정렬하고 상위 3명
exam.assign(test = np.where(exam['science'] >= 60, 'pass','fail')).sort_values('math').tail(3)
```

```python
m1 = mpg.copy()
# m1['total'] = m1['cty'] + m1['hwy']
m1 = m1.assign(total = m1['cty']+m1['hwy'])
m1.head(2)
```
