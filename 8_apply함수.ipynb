```python
# str(name)이 너무 길어서 변수명으로 빼고
# find를 이용해서 0과 -1인가 1과 -1? 값이 나오는데 그걸로 리턴값을 지정해주고
# 대소문자 구별을 위해서 lower를 쓰는게 효율적임 (데이터가 너무 많으면 언제 다 봄)
def get_categoty(name):
  name = str(name)
  if 'Chair' in name:
    return '의자'
  elif 'Table' in name:
    return '테이블'
  elif name.find('Mirror') != -1:
    return '거울'
  elif name.lower().find('sofa') != -1:
    return '소파'
  else :
    return 
  
df['카테고리'] = df['상품목록'].apply(get_categoty)
df
```
