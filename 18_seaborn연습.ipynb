```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
flights = sns.load_dataset('flights')
diamonds = sns.load_dataset('diamonds')
```

```python
# 막대 그래프
plt.figure()

# 기본값이 평균을 계산함 - 요일별 평균 팁
plt.subplot(2, 2, 1)
sns.barplot(data=tips, x='day', y='tip')
plt.title('요일별 팁')

plt.subplot(2, 2, 2)
sns.barplot(data=tips, x='day', y='tip', hue='sex')
plt.title('요일 별 평균 팁(성별)')

plt.subplot(2, 2, 3)
sns.barplot(data=tips, x='day', y='tip', estimator=sum)
plt.title('요일별 합계')

plt.subplot(2, 2, 4)
sns.barplot(data=tips, x='day', y='tip', palette='Set2')
plt.title('색상')

plt.show()
```

```python
# 개수 막대그래프
plt.figure(figsize=(12,8))

# 요일별 주문 수
plt.subplot(2, 2, 1)
sns.countplot(data=tips, x='day')
plt.title('요일별 주문 수')

# 요일별 성별 주문 수
plt.subplot(2, 2, 2)
sns.countplot(data=tips, x='day', hue='sex')

# 축 바꿈?
plt.subplot(2, 2, 3)
sns.countplot(data=tips, y='time')

# 요일별 흡연 여부에 따른 팁?주문?
plt.subplot(2, 2, 4)
sns.countplot(data=tips, x='day', hue='smoker', palette='viridis', order=['Sun','Sat','Fri','Thur']) # order은 x축 순서 직접 조정 가능

plt.show()
```

```python
# 박스플롯
plt.Figure(figsize=(12,8))

# 요일 별 주문 분포?
plt.subplot(2, 2, 1)
sns.boxplot(data=tips, x='day', y='total_bill')

# 성별로 기준?
plt.subplot(2, 2, 2)
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')

# 방향 바꾸기
plt.subplot(2, 2, 3)
sns.boxplot(data=tips, y='day', x='tip', orient='h') # 기본이 v (수직), h는 (수평)

plt.show()
```

```python
# 바이올린 차트
# 박스플롯보다 조금 더 자세하게 보고 싶을 때 사용
plt.Figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.violinplot(data=tips, x='day', y='total_bill')

plt.subplot(2, 2, 2)
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex')

plt.subplot(2, 2, 3)
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True)

plt.show()
```

```python
# 히스토그램
plt.figure(figsize=(12,8))

plt.subplot(2, 2, 1)
sns.histplot(data=tips, x='total_bill')

plt.subplot(2, 2, 2)
sns.histplot(data=tips, x='total_bill', hue='sex')

plt.subplot(2, 2, 3)
sns.histplot(data=tips, x='total_bill', kde=True) # 분포 곡선

plt.subplot(2, 2, 4)
sns.histplot(data=tips, x='total_bill', y='tip') # 2D?

plt.show()
```

```python
# 분포곡선 kde
plt.Figure(figsize=(12,8))

plt.subplot(2, 2, 1)
sns.kdeplot(data=tips, x='total_bill')

plt.subplot(2, 2, 2)
sns.kdeplot(data=tips, x='total_bill', hue='sex')

plt.subplot(2, 2, 3)
sns.kdeplot(data=tips, x='total_bill', y='tip')

plt.subplot(2, 2, 4)
sns.kdeplot(data=tips, x='total_bill',fill=True) # 색 채움?

plt.show()
```

```python
# 산점도
plt.Figure(figsize=(12,8))

plt.subplot(2, 2, 1)
sns.scatterplot(data=tips, x='total_bill', y='tip')

plt.subplot(2, 2, 2)
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')

plt.subplot(2, 2, 3)
sns.scatterplot(data=tips, x='total_bill', y='tip', size='size')

plt.subplot(2, 2, 4)
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', style='species', alpha=0.8) # species -> 종류별로

plt.show()
```

```python
# 선그래프
plt.figure(figsize=(12,8))

# 기본적으로 x축 별 값의 평균을 구함, 평균 선 기준으로 신뢰구간 밴드도 생성
plt.subplot(2, 2, 1)
sns.lineplot(data=flights, x='year', y='passengers')

plt.subplot(2, 2, 2)
sns.lineplot(data=flights, x='year', y='passengers',hue='month')

# 이것도 함수변경 가능 (sum이나 그런걸로)
plt.subplot(2, 2, 3)
sns.lineplot(data=tips, x='size', y='tip', hue='sex', style='smoker', estimator=sum, markers=True, errorbar=None)

plt.show()
```

```python
# 하트맵
plt.figure(figsize=(12,8))

plt.subplot(2, 2, 1)
corr = tips[['total_bill','tip','size']].corr()

sns.heatmap(corr, cmap='coolwarm', annot=True)

plt.subplot(2, 2, 2)
pivot = flights.pivot(
  index='month', columns='year', values='passengers'
)
sns.heatmap(pivot, cmap='YlOrRd')

plt.show()
```

```python
# Pairplot : 변수 간 관계를 한번에 시각화
g1 = sns.pairplot(tips[['total_bill','tip','size']])

g2 = sns.pairplot(
  data=tips, vars=['total_bill','tip','size'], hue='sex')

plt.show()
```

```python
# jointplot : 두 데이터 간 관계의 분포를 동시에 보여줌
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')

plt.show()
```

```python
plt.figure(figsize=(12,8))

# 기본
plt.subplot(2, 2, 1)
sns.barplot(data=tips, x='day', y='tip')
plt.title('기본 그래프')

plt.subplot(2, 2, 2)
sns.barplot(data=tips, x='day', y='tip', palette='viridis')
plt.title('그래프', fontsize=15, fontweight='bold')
plt.xlabel('요일')
plt.ylabel('팁')
plt.xticks(rotation=45) # 각도 돌리기 밑에 보면 글자가 45도 돌아감

plt.subplot(2, 2, 3)
sns.barplot(data=tips, x='day', y='tip', color='skyblue', edgecolor='navy', linewidth =2)
plt.grid(True, alpha=0.3)

plt.subplot(2, 2, 4)
sns.barplot(data=tips, x='day', y='tip', hue='sex')
plt.legend(title='성별')

plt.show()
```
