```python
import pandas as pd
cus = pd.read_csv('data/customers.csv')
pro = pd.read_csv('data/products.csv')
ord = pd.read_csv('data/orders.csv', parse_dates=['order_date'])

cus,pro,ord
```

```python
cus1 = pd.merge(cus, ord, on='user_id')
cus1
```

```python
cus1 = pd.merge(cus1,pro, on='product_id')
cus1
```

```python
# 총 매출액
cus1['total'] = cus1['unit_price'] * cus1['qty'] # 주문별 총 금액
cus1
cus1['total'].sum() # 총 매출액
```

```python
cus1.groupby('order_date')['total'].sum().reset_index(name='day_total').max() # 최고 매출 (일)
```

```python
cus1.groupby('order_date')['total'].sum().reset_index(name='day_total').min() # 최저 매출 (일)
```
