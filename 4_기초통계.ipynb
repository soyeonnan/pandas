```python
# 컬럼명 변경하기
# mpg의 cty -> city, hwy -> highway로 변경
# 처음 가져 온 원본 데이터를 직접적으로 변형하면서 작업하지 않음
mpg_rename = mpg.copy() # 복사본 생성
mpg_rename.rename(columns={'cty':'city','hwy':'highway'},inplace=True)
mpg_rename
```

```python
# 컬럼 추가하기 ( 파생 변수 생성 )
# 기존 컬럼(변수)를 이용해서 만들어진 새 컬럼을 파생 변수라고 함 
df = pd.DataFrame({
  'var1':[4,3,8],
  'var2':[2,6,1]
})
df
```

```python
df['sum'] = df['var1'] + df['var2']
df
```

```python
df['mean'] = df['sum'] / 2
df
```

```python
# mpg데이터 프레임에서 cty와 hwy의 평균값을 나타내는 total 컬럼을 추가
mpg['total'] =( mpg['cty'] + mpg['hwy']) /2
mpg
```

```python
#  total이 20이상이면 pass, 아니면 fail
# 넘파이에 where이라는 함수를 이용해서 처리가 가능함
import numpy as np

mpg['test'] = np.where(mpg['total'] >= 20, 'pass','fail')
mpg.head()
```

```python
# 해당 컬럼에 값이 종류별 몇 개씩 있는지
mpg['test'].value_counts()
```

```python
# 조건을 여러 개
# total이 30이면 A, 20이상이면 B, 나머지는 C
mpg['grade'] = np.where(mpg['total'] >= 30, 'A',np.where(mpg['total'] >= 20 ,'B','C'))
mpg
```

```python
mpg['grade'].value_counts()
```

```python
mpg['category'].value_counts()
```

```python
# category가 compact, subcompact, 2seater면 small
# 아니면 large로 표시하는 size 파생 변수를 추가 하십쇼
# where 함수에서는 and, or 사용 못 함 -> and : &, or : |로 사용
mpg['size'] = np.where((mpg['category'] == 'compact') |
                       (mpg['category'] == 'subcompact')|
                       (mpg['category'] == '2seater'),'small' , 'large')

mpg['size'] = np.where(
  mpg['category'].isin(['compact','subcompact','2seater']),'small','large'
)

mpg['size'].value_counts()
```

```python
# midwest.csv파일을 가져와서 데이터 구조 파악
# poptotal(전체인구)를 total로,
# popasian(아시아인구)를 asian으로 변경
# total과 asian을 이용해서 전체 인구 대비 아시아 인구 비율
# 파생변수(ratio)추가
# '아시아 인구 비율'이 '아시아 인구 비율 평균 '을 초과하면 large,
# 아니면 small을 나타내틑 파생변수(size) 추가
mid = pd.read_csv('data/midwest.csv')
mid.head()
# print(mid.tail())
# print(mid.shape)
# print(mid.info())
mid.columns
```

```python
_mid = mid.copy()
_mid.rename(columns={'poptotal':'total','popasian':'asian'},inplace=True)
_mid.columns
```

```python
_mid['ratio'] = (_mid['asian'] / _mid['total']) * 100
_mid['ratio'].value_counts()
_mid.loc[:2],['country','total','asian','ratio']
```

```python
avg = _mid['ratio'].mean()
_mid['size'] = np.where(_mid['ratio'] > avg, 'large','small')
_mid['size'].value_counts()
```
